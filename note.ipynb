# **BERT**

### **What is BERT ?**
*BERT is a pretrained language model that has been trained on large amounts of text to understand English language patterns, grammar, context, and relationships between words. When we give BERT some text, it produces contextual embeddings for each token, meaning that the embedding of a word depends on the words around it. These embeddings capture richer semantic meaning than traditional word embeddings.*

- BERT is a pretrained language model that trained to understand language(English)
- Like we train CNN on imaegs and get prediction for unseen Images, **Bert** also a LM(language Model) that train on billions of words and sentences to understand
   - grammar
   - context
   - word relationships
   - language patterns
- After training, when you give a new unseen text, *"Even if BERT never saw this exact sentence before, it can still generate meaningful embeddings because it learned general language patterns"*
- If we give an input text It outputs
  - **Contextual embeddings (hidden states)**
  - **CLS embedding (special token)**

### **What BERT is trained for (Pretraining stage)**
BERT is trained on huge text data using
- (1) Masked Language Model (MLM)
     - “I love playing [MASK]” >>> Some words are hidden BERT predicts(football, games, etc.), Learns word meaning from context
- (2) Next Sentence Prediction (older BERT)
     - Given :  A: “I went to the bank to withdraw money.”, B: “The cashier gave me cash.” >>> BERT learns : Are sentences related or not ?

### **Bert output ?**

If we give an input text It outputs :
  - **Contextual embeddings (hidden states)**
     - For every token in your sentence, BERT gives a vector(One vector per word/token)
     - “I love AI” >>> “I” → vector (768 numbers), “love” → vector (768 numbers), “AI” → vector (768 numbers)
  - **CLS embedding (special token)**
    - BERT also adds a special token => [CLS] = summary of the whole sentence
    - We also get > CLS → one 768-dimensional vector (This is NOT a label. It is just a meaning representation)
      
#### What is CLS token ?
- It is a summary of the whole sentence/article (compressed meaning of the entire news article)
- We use this for classification

### **What “fine-tuning” actually means**

#### **Before Fine-Tuning we download a pretrained BERT**
- This BERT already knows :
   - English grammar
   - Word meanings
   - Context
   - Sentence structure
- For example:
    - "The team won the match."
    - BERT might create a CLS embedding like : [0.21, -0.44, 0.91, ...]
    - And for : "The parliament passed a bill." : [-0.12, 0.77, -0.31, ...]
- These vectors contain meaning, but BERT doesn't know:
    - Vector A = Category A
    - Vector B = Category B
    - It only knows language

#### **What Training Does**
- Suppose the correct answer is: Sports
- The loss function calculates:

*"You should have predicted Sports, but you didn't."*

Then backpropagation starts

#### **What Gets Updated?**
- **Option 1: Train only the classifier**
  - BERT = frozen
  - Classifier = trainable
  - **Only the new classification layer learns, BERT stays exactly as it was downloaded.**
- **Option 2: Fine-tune BERT (most common)**
  - BERT = trainable
  - Classifier = trainable
  - ****
    
#### **What Does "BERT weights slightly adjust" Mean ?**
- Imagine BERT already understands: football, goal, stadium, player
- These words already have good embeddings
- But your dataset may reveal : Whenever these concepts appear, the article is usually Sports
- So training gently modifies BERT's internal parameters.
   - NOT : Complete rewrite of language knowledge
   - BUT : Small adaptation toward Out task
#### **Why "Slightly"?**
- Because pretrained BERT already spent enormous compute learning language
- If we change everything aggressively : BERT forgets language knowledge (This is called catastrophic forgetting)
- So fine-tuning usually uses a very small learning rate
- The model makes tiny adjustments
*Human Example>br>
Imagine a lawyer who already speaks perfect English. Pretraining: Learn English for 20 years , Now they want to become a sports journalist. Fine-tuning: Learn sports reporting, They don't relearn English. They adapt existing knowledge. That's exactly what fine-tuning does.*

#### **What Happens Internally ?**
- Before fine-tuning:
  - Sports article embedding
  - Politics article embedding
    
     may not be well separated

- After training:
    - Sports embeddings cluster together
    - Politics embeddings cluster together
    - Business embeddings cluster together
    
     The model reshapes its embedding space so classification becomes easier.Now gradients flow through the entire network.
